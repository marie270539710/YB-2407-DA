**Time Series Plot: Fish Population Over Time**

This will visualize how the fish population changes over time at each site.

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Sample data
data = {
    'Site ID': ['AV-1', 'AV-2', 'AV-3', 'AV-1', 'AV-2', 'AV-3'],
    'Date': ['2025-10-25', '2025-10-25', '2025-10-25', '2025-11-15', '2025-11-15', '2025-11-15'],
    'Species': ['Brown Trout', 'Shortfin Eel', 'Freshwater Shrimp', 'Rainbow Trout', 'Kokopu', 'No fish observed'],
    'Count': [50, 75, 120, 35, 40, None],
    'Average Size (cm)': [32, 45, 2, 28, 18, None],
    'Temperature (Â°C)': [15.2, 14.8, 13.6, 12.4, 11.9, 10.8],
    'pH': [7.8, 7.5, 7.3, 7.6, 7.4, 7.2],
    'Dissolved Oxygen (mg/L)': [8.5, 7.9, 6.8, 8.2, 7.7, 6.2]
}

# Convert the data into a pandas DataFrame
df = pd.DataFrame(data)

# Convert the 'Date' column to datetime format
df['Date'] = pd.to_datetime(df['Date'])

# Plotting Time Series of Fish Population
plt.figure(figsize=(10, 6))
sns.lineplot(data=df, x='Date', y='Count', hue='Site ID', marker='o')
plt.title('Fish Population Over Time by Site')
plt.xlabel('Date')
plt.ylabel('Fish Count')
plt.xticks(rotation=45)
plt.legend(title='Site ID')
plt.tight_layout()
plt.show()
```

**2. Correlation Heatmap: Fish Count vs Water Quality Parameters**

This will show the relationship between fish count, average size, and water quality parameters (temperature, pH, dissolved oxygen).

```python
# Calculating the correlation matrix for the relevant columns
correlation_data = df[['Count', 'Average Size (cm)', 'Temperature (Â°C)', 'pH', 'Dissolved Oxygen (mg/L)']].corr()

# Plotting the correlation heatmap
plt.figure(figsize=(8, 6))
sns.heatmap(correlation_data, annot=True, cmap='coolwarm', fmt='.2f', linewidths=0.5)
plt.title('Correlation Heatmap: Fish Count and Water Quality')
plt.tight_layout()
plt.show()
```

**3. Boxplot: Fish Population Count Across Sites**

This boxplot will compare the distribution of fish counts across the three sites, helping us identify potential trends or outliers.

```python
# Plotting the boxplot for fish counts across sites
plt.figure(figsize=(10, 6))
sns.boxplot(x='Site ID', y='Count', data=df)
plt.title('Fish Count Distribution Across Sites')
plt.xlabel('Site ID')
plt.ylabel('Fish Count')
plt.tight_layout()
plt.show()
```
